```python
import keras # machine learning lib
import numpy as np #do some math
import pandas as pd #read in our data
import matplotlib.pyplot as plt # graph our model
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('drive/My Drive/IowaHousingPrices.csv')
```

```python
SquareFeet= df[['SquareFeet']].values
SalePrice= df[['SalePrice']].values
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(1, input_shape=(1, )))
model.compile(keras.optimizers.Adam(lr=1),'mean_squared_error')
```

```python
model.fit(SquareFeet,SalePrice, epochs = 30, batch_size=10)
```

```python
df.plot(kind= 'scatter', x= 'SquareFeet', y='SalePrice', title="HP", )
y_pred = model.predict(SquareFeet)

plt.plot(SquareFeet, y_pred, color= 'red')
```
